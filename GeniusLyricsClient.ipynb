```python
dict1 = {}
```

```python
with open("soy_songs.csv", "rt") as infile:
    reader = csv.reader(infile)
    headers = next(reader)[1:]
    for row in reader:
        dict1[row[0]] = {key: value for key, value in zip(headers, row[1:])}
```

```python
dict1
```

```python
for key in dict1:
#     if int(dict1[key]['year']) <=2021 or int(dict1[key]['year']) >= 2024 or dict1[key]['wasWinner'] == 'yes':
#         continue
    if 'lyrics' in dict1[key]:
        continue
    else:
        dict1[key]['lyrics']=''
        continue
    artist_name = dict1[key]['artist']
    artist_name = artist_name.split(' featuring')[0]
    artist_name = artist_name.split(' &')[0]
    artist_name = artist_name.split(' (')[0]
    artist_name = artist_name.split(' with')[0]
    if artist_name == 'Paul McCartney':
        artist_name = 'Stevie Wonder'

    try:
        artist_genius = genius.search_artist(artist_name,max_songs=1)
    except:
        time.sleep(30)
        artist_genius = genius.search_artist(artist_name,max_songs=1)
    if artist_genius == None:
        continue
    try:
        song = artist_genius.song(dict1[key]['song'])
    except:
        time.sleep(30)
        song = artist_genius.song(dict1[key]['song'])
    if song == None:
        continue
    lyrics = song.lyrics.replace('\n',' ')
    dict1[key]['lyrics'] = lyrics
```

```python
with open('soy_lyrics.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in dict1.items():
       writer.writerow([key, value])
```
