```python
import csv
import spacy
```

```python
import re
```

```python
regex = re.compile('[^a-zA-Z\s]')
```

```python
dict1 = {}
```

```python
with open("best_rap_song.csv", "rt") as infile:
    reader = csv.reader(infile)
    headers = next(reader)[1:]
    for row in reader:
        dict1[row[0]] = {key: value for key, value in zip(headers, row[1:])}
```

```python
from monkeylearn import MonkeyLearn

ml = MonkeyLearn('48d44c41482e3316b67f34dd70b7633c27394a57')

model_id = 'cl_NDBChtr7'
```

```python
for key in dict1:
    dict1[key]['lyrics'] = dict1[key]['lyrics'].lower()
    x = dict1[key]['lyrics'].find(']')
    dict1[key]['lyrics'] = dict1[key]['lyrics'][x+1:]
#     dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('contributors','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('[chorus]','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('pre chorus','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('verse','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('embed','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('[chorus ]','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('[instrumental bridge]','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('[outro]','')
    dict1[key]['lyrics'] = dict1[key]['lyrics'].replace('lyrics','')
    dict1[key]['lyrics'] = regex.sub('', dict1[key]['lyrics'])
    
```
